# CDF & PDF on MBA Class Placement Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Placement_Data_Full_Class.csv")
df.head()
```

```python
plt.hist(df['ssc_p'], bins=10, density=True)
plt.xlabel("secondary school percentage")
plt.ylabel("Frequency")
plt.show()
plt.hist(df['hsc_p'], bins=10)
plt.xlabel("higher school percentage")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

Showing the  PDF using 'hsc_p' and 'ssc_p' column

```python
sns.histplot(
    df['ssc_p'], kde=True,
    stat="density", kde_kws=dict(cut=3)
).set(title="Probability Density Function")
plt.xlabel('secondary school percentage')
plt.show()
sns.histplot(
    df['hsc_p'], kde=True,
    stat="density", kde_kws=dict(cut=3)
).set(title="Probability Density Function")
plt.xlabel('higher secondary school percentage')
plt.show()
```

Showing the CDF using 'hsc_p' and 'ssc_p' column

```python
sns.ecdfplot(
    df['ssc_p'], 
    stat='proportion'
).set(title="Cummulative Density Function")
plt.xlabel('secondary school percentage')
plt.show()
sns.ecdfplot(
    df['hsc_p'], 
    stat='proportion'
).set(title="Cummulative Density Function")
plt.xlabel('higher secondary school percentage')
plt.show()
```
